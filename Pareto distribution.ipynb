# Pareto distribution - The "80-20 Rule"

A Pareto continuous random variable.

### PDF
P(x,b) = b*(k^b)/x^(b+1)

k is minimum value of x , b is a shaping parameter

In scipy k=1.

PDF looks like = b/x^(b+1)

```python
from scipy.stats import pareto
import matplotlib.pyplot as plt
import numpy as np
```

## Plotting pareto distribution

```python
b = 2.62
fig, ax = plt.subplots(1, 1)
x = np.linspace(pareto.ppf(0.01, b), pareto.ppf(0.99, b), 100)
ax.plot(x, pareto.pdf(x, b), 'r-', lw=5, alpha=0.6, label='pareto pdf')
```

## Generating sample and plotting

```python
r = pareto.rvs(b, size=1000)
import seaborn as sns
sns.distplot(r,kde=True)
plt.show()
```
